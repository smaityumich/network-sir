```python
import numpy as np
from baler_sir import *
import itertools
```

```python
n = 1000
p = 0.001
nodes = np.zeros(n)
nodes[:3] = 1
```

```python
adj = dict()
for i, _ in enumerate(nodes):
    adj[i] = []
    
for i in range(1, n):
    for j in range(i):
        if np.random.binomial(1, p):
            adj[i].append(j)
            adj[j].append(i)
            
adj
```

```python
graph = Graph(nodes, adj_matrix=adj)
```

```python
for _ in range(100):
    graph.recovery_step()
    graph.infection_step()
    i, r = graph.current_status()
    print(f'Infected {i}, recovered {r}\n')
```
